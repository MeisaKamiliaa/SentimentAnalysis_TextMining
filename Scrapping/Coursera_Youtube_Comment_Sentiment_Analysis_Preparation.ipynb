## Import Data

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd

df = pd.read_csv('comment_data.tsv', sep='\t')
df
```

## Extract Context Based On The Keyword

### Define keyword

```python
keyword = {
    'coursera',
    'online course',
    'platform',
    'course',
    'learning'
}
```

### Context extraction

```python
df = df[df['comment'].str.contains('|'.join(keyword), case=False, na=False)]
```

### Save the context in a sentiment dataframe

```python
sentiment_data = df['comment']
sentiment_data
```

### Save into TSV

```python
sentiment_data.to_csv('sentiment_data.tsv', index=False)
```
